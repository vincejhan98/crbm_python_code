```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Bivariate Gaussian distribution
correlation = 0.5
covariance = [[1, correlation], [correlation,1]]

# regular sampling, (such as prior sampling)
num_samples = 5000
regular_samples = np.random.multivariate_normal([0,0], covariance, num_samples)
```

```python
# Gibbs Sampling
x_samples = [1]
y_samples = [1]

for _ in range(num_samples):
    recent_y = y_samples[-1] # most recent y value
    new_x = np.random.normal(recent_y * correlation, np.sqrt(1-correlation*correlation)) # x[i+1] ~ P(x | y = y[i])
    new_y = np.random.normal(new_x * correlation, np.sqrt(1-correlation*correlation)) # y[i+1] ~ P(y | x = x[i+1])
    x_samples.append(new_x)
    y_samples.append(new_y)
```

```python
plt.scatter(regular_samples[:,0], regular_samples[:,1])
plt.scatter(x_samples, y_samples)
```
